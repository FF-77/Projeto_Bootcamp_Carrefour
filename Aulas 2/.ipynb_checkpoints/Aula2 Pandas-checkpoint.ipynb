```python
#Importando Pandas

import pandas as pd
```

```python
#Lewitura dos arquivos

df1 = pd.read_excel('C:/Users/Francisco Franco/Desktop/Bootcamp/Aulas Python Pandas/Aulas 2/Aracaju.xlsx')
df2 = pd.read_excel('C:/Users/Francisco Franco/Desktop/Bootcamp/Aulas Python Pandas/Aulas 2/Fortaleza.xlsx')
df3 = pd.read_excel('C:/Users/Francisco Franco/Desktop/Bootcamp/Aulas Python Pandas/Aulas 2/Natal.xlsx')
df4 = pd.read_excel('C:/Users/Francisco Franco/Desktop/Bootcamp/Aulas Python Pandas/Aulas 2/Recife.xlsx')
df5 = pd.read_excel('C:/Users/Francisco Franco/Desktop/Bootcamp/Aulas Python Pandas/Aulas 2/Salvador.xlsx')
```

```python
#Juntando todos os arquivos

df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#Mostrando as 5 primeiras linhas
df.head()
```

```python
#Mostrando as 5 últimas linhas
df.tail()
```

```python
#Verificando o tipo de cada coluna
df.dtypes
```

```python
#Alterando tipo de dado
df['LojaID'] = df['LojaID'].astype("object")
```

```python
df.dtypes
```

```python
#Consultar linhas com valores faltantes
df.isnull().sum()
```

```python
#Substituindo os valores nnulos pela média
df['Vendas'].fillna(df['Vendas'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
#Mostra aleatória
df.sample(15)
```

```python
#Substituindo valores nulos por zero
df['Vendas'].fillna(0, inplace=True)
```

```python
df
```

```python
#Apagando as linhas com valores nulos
df.dropna(inplace=True)
```

```python
#Apagando as linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=['Vendas'], inplace=True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how='all', inplace=True)
```

```python
#Criando coluna de receita
df['Receita'] = df['Vendas'] * df['Qtde']

df
```

```python
#Achar a quantidade vendinda
df['Receita/Vendas'] = df['Receita'] / df['Vendas']
df.head()
```

```python
#Retorna o maior valor
df['Receita'].max()
```

```python
#Retornar o menor valor
df['Receita'].min()
```

```python
#Verificar os tops de vendas metodo nlargest
df.nlargest(3, 'Receita')
```

```python
#Verificar os que menos venderam | metodo nsmallest
df.nsmallest(3, 'Receita')
```

```python
#Agrupamento por cidade
df.groupby("Cidade")['Receita'].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values('Receita', ascending=False).head(10)
```

TRABALHANDO COM DATAS

```python
#Transformando a coluna de data em tipo inteiro
df['Data'] = df['Data'].astype('int64')
df.dtypes
```

```python
#Transformando a coluna de data em data
df['Data'] = pd.to_datetime(df['Data'])
df.dtypes
```

```python
#Agrupamento por ano
df.groupby(df['Data'].dt.year)['Receita'].sum()
```

```python
#Criando uma nova coluna com o ano
df['Ano_Venda'] = df['Data'].dt.year

df
```

```python
#Extraindo o mês e o dia
df['Mês_Venda'], df['Dia_Venda'] = (df['Data'].dt.month, df['Data'].dt.day)
df
```

```python
#Retornando a data mais antiga
df['Data'].min()
```

```python
#Calculando diferença em dias
df['Diferença_dia'] = df['Data'] - df['Data'].min()
df
```

```python
#Criando a coluna de trimestre
df['Trimestre_Venda'] = df['Data'].dt.quarter
df.tail(20)
```

```python
#Filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df['Data'].dt.year == 2019) & (df['Data'].dt.month == 3)]
vendas_marco_19
```

```python
#Qual a receita total
df12['Valor Venda'].sum()
```

```python
#Qual o custo total
df12['custo'] = df12['Custo Unitário'].mul(df12['Quantidade']) #Criando coluna de custo
```

```python
df12.head(1)
```

```python
#Soma do custo total
df12['custo'].sum()
```

```python
#Soma do custo total arredondando para 2 casas decimais
round(df12['custo'].sum(), 2)
```

```python
#Temos a receita e o custo agora vamos cacular o lucro
#vamos criar a coluna lucro
df12['lucro'] = df12['Valor Venda'] - df12['custo']

df12['lucro'].sum()
```

```python
#criando uma coluna com total de dias para enviar produto
df12['Tempo envio'] = df12['Data Envio'] - df12['Data Venda']


df12['Tempo envio']
```

AGORA QUEREMOS VER O TEMPO DE ENVIO PARA CADA MARCA, E PARA ISSO PRECISAMOS TRANSFORMA A COLUNA TEMPO ENVIO EM NÚMERICO

```python
#Extraindo apenas os dias
df12['Tempo envio'] = (df12['Data Envio'] - df12['Data Venda']).dt.days
df12['Tempo envio'] 
```

```python
#Verificando o tipo de coluna Tempo envio
df12['Tempo envio'].dtype
```

```python
#Média do tempo de envio por Marca
df12.groupby('Marca')['Tempo envio'].mean()
```
